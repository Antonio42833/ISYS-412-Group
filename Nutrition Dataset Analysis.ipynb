## Calorie Consumption by Meal Type

```python
meal_data = DF.groupby('Meal_Type')['Calories (kcal)'].sum().reset_index()

fig = go.Figure(data=[go.Bar(name='Calories', x=meal_data['Meal_Type'], y=meal_data['Calories (kcal)'])])

fig.update_layout(
    title='Calorie Consumption by Meal Type',
    xaxis_title='Meal Type',
    yaxis_title='Calories (kcal)',
    barmode='stack')

fig.show()
```

## Calorie Consumption by Category

```python
category_data = DF.groupby('Category')['Calories (kcal)'].sum().reset_index()

fig = go.Figure(data=[
    go.Bar(
        name='Calories',
        x=category_data['Category'], 
        y=category_data['Calories (kcal)'],
        marker=dict(color='blue'))])

fig.update_layout(
    title='Calorie Consumption by Category',
    xaxis_title='Category',
    yaxis_title='Calories (kcal)',
    template='plotly')

fig.show()
```

## Calorie and Water Intake by person

```python
import plotly.graph_objects as go

# Group by User_ID and calculate total Calories and Water Intake
user_data = DF.groupby('User_ID')[['Calories (kcal)', 'Water_Intake (ml)']].sum().reset_index()

# Create a scatter plot
fig = go.Figure()

fig.add_trace(
    go.Scatter(
        x=user_data['Calories (kcal)'],
        y=user_data['Water_Intake (ml)'],
        mode='markers',
        marker=dict(size=12, color='purple', opacity=0.8),
        text=user_data['User_ID'],  # Display User_ID on hover
    )
)

fig.update_layout(
    title='Calorie Consumption vs Water Intake by User',
    xaxis_title='Total Calories (kcal)',
    yaxis_title='Total Water Intake (ml)',
    template='plotly_dark',  # Optional: dark styling template
    showlegend=False
)

fig.show()
```

## Protein vs. Calorie Intake by food item

```python
import plotly.graph_objects as go

# Group by Food_Item and sum Calories and Protein
food_data = DF.groupby('Food_Item')[['Calories (kcal)', 'Protein (g)']].sum().reset_index()

# Create a scatter plot
fig = go.Figure()

fig.add_trace(
    go.Scatter(
        x=food_data['Calories (kcal)'],
        y=food_data['Protein (g)'],
        mode='markers',
        marker=dict(size=10, color='green', opacity=0.7),
        text=food_data['Food_Item'],  # Hover text with food item names
    )
)

fig.update_layout(
    title='Protein vs Calorie Intake by Food Item',
    xaxis_title='Calories (kcal)',
    yaxis_title='Protein (g)',
    template='plotly_dark'
)

fig.show()
```

## Correlation Matrix Heatmap

```python
import seaborn as sns
import matplotlib.pyplot as plt

correlation_matrix = DF[['Calories (kcal)', 'Protein (g)', 'Carbohydrates (g)', 'Fat (g)', 'Fiber (g)', 'Sugars (g)', 'Sodium (mg)', 'Cholesterol (mg)', 'Water_Intake (ml)']].corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix Heatmap')
plt.show()
```

## Nutritional Profile by Meal_Type

```python
from plotly.graph_objects import Figure, Scatterpolar

radar_data = DF.groupby('Meal_Type')[['Protein (g)', 'Carbohydrates (g)', 'Fat (g)', 'Fiber (g)']].mean().reset_index()

fig = go.Figure()
for i, meal in radar_data.iterrows():
    fig.add_trace(
        go.Scatterpolar(
            r=meal[1:].tolist(),
            theta=['Protein', 'Carbohydrates', 'Fat', 'Fiber'],
            fill='toself',
            name=meal['Meal_Type']
        )
    )
fig.update_layout(title='Nutritional Profile by Meal Type', polar=dict(radialaxis=dict(visible=True)))
fig.show()
```

## Calorie Contribution of Macronutrients

```python
macronutrient_calories = DF.groupby('Category')[['Protein (g)', 'Carbohydrates (g)', 'Fat (g)']].sum().reset_index()
macronutrient_calories['Protein Calories'] = macronutrient_calories['Protein (g)'] * 4
macronutrient_calories['Carbohydrate Calories'] = macronutrient_calories['Carbohydrates (g)'] * 4
macronutrient_calories['Fat Calories'] = macronutrient_calories['Fat (g)'] * 9

fig = go.Figure(data=[
    go.Bar(name='Protein Calories', x=macronutrient_calories['Category'], y=macronutrient_calories['Protein Calories']),
    go.Bar(name='Carbohydrate Calories', x=macronutrient_calories['Category'], y=macronutrient_calories['Carbohydrate Calories']),
    go.Bar(name='Fat Calories', x=macronutrient_calories['Category'], y=macronutrient_calories['Fat Calories']),
])
fig.update_layout(title='Calorie Contribution of Macronutrients', barmode='stack')
fig.show()
```

## when most people consume the highest number of calories based on weekdays

```python
DF['Date'] = pd.to_datetime(DF['Date'])  # Ensure Date column is in datetime format
DF['Weekday'] = DF['Date'].dt.day_name()  # Add a new column for the weekday
```

```python
weekday_calories = DF.groupby('Weekday')['Calories (kcal)'].sum().reindex(
    ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
).reset_index()

print(weekday_calories)
```

```python
import plotly.express as px

fig = px.bar(
    weekday_calories,
    x='Weekday',
    y='Calories (kcal)',
    title='Calorie Consumption by Weekday',
    labels={'Calories (kcal)': 'Total Calories'},
    color='Calories (kcal)',  # Optional: Add color for visual flair
    color_continuous_scale='Blues'
)

fig.show()
```
