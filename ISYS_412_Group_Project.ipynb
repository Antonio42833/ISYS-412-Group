##### What nutritional patterns can be discovered among individual foods, and how can we identify foods that are most and least aligned with general dietary health guidelines?

```python
import matplotlib.pyplot as plt
import pandas as pd

df1 = pd.read_csv('FOOD-DATA-GROUP1.csv')
df2 = pd.read_csv('FOOD-DATA-GROUP2.csv')
df3 = pd.read_csv('FOOD-DATA-GROUP3.csv')
df4 = pd.read_csv('FOOD-DATA-GROUP4.csv')
df5 = pd.read_csv('FOOD-DATA-GROUP5.csv')

df = pd.concat([df1, df2, df3 ,df4 ,df5], ignore_index=True)

df
```

## Which foods have the highest and lowest calorie content per 100g?

### Highest

```python
df.sort_values(by='Caloric Value', ascending=False).head(20)
```

```python
top20 = df.sort_values(by='Caloric Value', ascending=False).head(20)

plt.barh(top20['food'], top20['Caloric Value'])
plt.xlabel('Calories per 100g')
plt.title('Top 20 Highest-Calorie Foods')
plt.gca().invert_yaxis()
plt.show()
```

### Lowest

```python
df.sort_values(by='Caloric Value', ascending=True).head(20)
```

```python
bottom20 = df.sort_values(by='Caloric Value', ascending=True).head(20)

plt.barh(bottom20['food'], bottom20['Caloric Value'])
plt.xlabel('Calories per 100g')
plt.title('Top 20 Lowest-Calorie Foods')
plt.gca().invert_yaxis()
plt.show()

# It was more interesting seeing the zero calorie foods instead of 20, 1 calorie foods.
```

## Which foods are highest and lowest in protein relative to calories (most protein-dense)?

### Highest

```python
df_filtered = df[df['Caloric Value'] > 0]

df_filtered['Protein Density'] = df_filtered['Protein'] / df_filtered['Caloric Value']

most_protein_dense = df_filtered.sort_values(by='Protein Density', ascending=False)

most_protein_dense[['food', 'Protein', 'Caloric Value', 'Protein Density']].head(20)
```

```python
top_protein_dense = most_protein_dense.head(20)

plt.barh(top_protein_dense['food'], top_protein_dense['Protein Density'])
plt.xlabel('Protein per Calorie')
plt.title('Top 10 Protein-Dense Foods')
plt.gca().invert_yaxis()
plt.show()
```

### Lowest

```python
# I did this so we don't see all the zeros

df_filtered2 = df[(df['Caloric Value'] > 0) & (df['Protein'] > 0)].copy()

df_filtered2['Protein Density'] = df_filtered2['Protein'] / df_filtered2['Caloric Value']

least_protein_dense = df_filtered2.sort_values(by='Protein Density', ascending=True)

least_protein_dense[['food', 'Protein', 'Caloric Value', 'Protein Density']].head(20)
```

```python
bottom_protein_dense = least_protein_dense.head(20)

import matplotlib.pyplot as plt

plt.barh(bottom_protein_dense['food'], bottom_protein_dense['Protein Density'])
plt.xlabel('Protein per Calorie')
plt.title('Top 20 Least Protein-Dense Foods')
plt.gca().invert_yaxis()
plt.show()
```

&nbsp;&nbsp;&nbsp;Are there foods that are both low in fat and high in fiber? discover natural categories (e.g., energy-dense vs. nutrient-dense foods)?

```python
# Define thresholds (you can adjust these based on your dataset)
low_fat_threshold = 3     # grams of fat per 100g (example: "low fat" definition)
high_fiber_threshold = 5  # grams of fiber per 100g

# Filter for foods that meet both criteria
low_fat_high_fiber = df[
    (df['Fat'] <= low_fat_threshold) & 
    (df['Dietary Fiber'] >= high_fiber_threshold)
]

# Display results
low_fat_high_fiber[['food', 'Fat', 'Dietary Fiber']]
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans

# Select relevant features
features = df[['Caloric Value', 'Protein', 'Fat', 'Carbohydrates', 'Dietary Fiber']].dropna()

# Standardize data
scaler = StandardScaler()
X_scaled = scaler.fit_transform(features)

# Fit K-Means (let's say 3 clusters to start)
kmeans = KMeans(n_clusters=3, random_state=42)
df['Cluster'] = kmeans.fit_predict(X_scaled)

# Visualize clusters
import seaborn as sns
import matplotlib.pyplot as plt

sns.scatterplot(data=df, x='Caloric Value', y='Protein', hue='Cluster', palette='Set2')
plt.title('Clusters of Foods by Calories and Protein')
plt.show()
```

&nbsp;&nbsp;&nbsp;Which foods have extremely high sugar or sodium content per serving?

```python
# Drop NaNs for safe filtering
df_clean = df[['food', 'Sugars', 'Sodium']].dropna()

# Calculate 95th percentile for sugar and sodium
sugar_threshold = df_clean['Sugars'].quantile(0.95)
sodium_threshold = df_clean['Sodium'].quantile(0.95)

# Filter foods above those thresholds
high_sugar = df_clean[df_clean['Sugars'] >= sugar_threshold]
high_sodium = df_clean[df_clean['Sodium'] >= sodium_threshold]

# Optional: Top 10 from each
high_sugar_top10 = high_sugar.sort_values(by='Sugars', ascending=False).head(10)
high_sodium_top10 = high_sodium.sort_values(by='Sodium', ascending=False).head(10)
```

```python
import matplotlib.pyplot as plt

# Plot top sugar-heavy foods
plt.barh(high_sugar_top10['food'], high_sugar_top10['Sugars'])
plt.xlabel('Sugars (g)')
plt.title('Top 10 Foods by Sugar Content per 100g')
plt.gca().invert_yaxis()
plt.show()

# Plot top sodium-heavy foods
plt.barh(high_sodium_top10['food'], high_sodium_top10['Sodium'])
plt.xlabel('Sodium (mg)')
plt.title('Top 10 Foods by Sodium Content per 100g')
plt.gca().invert_yaxis()
plt.show()
```
